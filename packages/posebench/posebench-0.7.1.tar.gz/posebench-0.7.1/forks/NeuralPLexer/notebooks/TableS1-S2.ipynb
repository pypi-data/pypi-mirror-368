```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_df(ds_name, mode, sampler="langevin_simulated_annealing"):
    dfs = []
    try:
        df = pd.read_csv(f"../neuralplexermodels_downstream_datasets_predictions/predictions/{ds_name}/set0/trainfrac1/default/af2_template/{mode}/{sampler}_T1.0_NFE100/summary.csv")
        df['group_id'] = 0
        dfs.append(df)
    except:
        pass
    for i in range(1, 6):
        dfi = pd.read_csv(f"../neuralplexermodels_downstream_datasets_predictions/predictions/{ds_name}/msa512_set1to5/default/af2_template_msa512/{mode}/{sampler}_T1.0_NFE100/model{i}/summary.csv")
        dfi['group_id'] = i
        dfs.append(dfi)
        print(len(dfi))
    return pd.concat(dfs)
```

```python
from IPython.display import display, HTML

def pretty_print(df):
    return display( HTML( df.to_html().replace("\\n","<br>") ) )

df = get_df("pocketminer", "inferred_prior")
df_af2 = df[['sample_id', 'TMScore_template_ref', 'RMSD_template_ref']]
df_af2['TM-score'] = df_af2['TMScore_template_ref']
df_af2['Backbone RMSD'] = df_af2['RMSD_template_ref']
#df_af2['lDDT-BS'] = df_af2['lDDT-BS_template_ref']
df_af2['Ligand RMSD'] = ''
del df_af2['TMScore_template_ref']
del df_af2['RMSD_template_ref']
#del df_af2['lDDT-BS_template_ref']
df_af2_summary = df_af2.describe()
df_af2_summary['sampler'] = '-'
df_af2_summary['Method'] = 'AlphaFold2'
df_af2_summary = df_af2_summary[['Method', 'sampler', 'TM-score', 'Backbone RMSD']]
df_af2_summary_top1 = df_af2.sort_values('TM-score', ascending=False).drop_duplicates('sample_id').describe()
df_af2_summary_top1

df_af2_summary = df_af2_summary.join(df_af2_summary_top1, lsuffix='\\n (all predictions)', rsuffix='\\n (per-target top-1)')
df_af2_summary = df_af2_summary.drop(['count', 'min', 'max'])

all_summary = [df_af2_summary]
sampler_name = {'DDIM': 'DDIM', 'langevin_simulated_annealing': 'LSA-SDE'}
mode_name = {'discard_ligand': 'NeuralPLexer \\n (no ligand)', 'inferred_prior': 'NeuralPLexer \\n (ours)'}

for mode in ['discard_ligand', 'inferred_prior']:
    for sampler in ['DDIM', 'langevin_simulated_annealing']:
        df = get_df("pocketminer", mode, sampler=sampler)
        df_nplx = df[['sample_id', 'TMScore', 'Calpha_rmsd', 'ligand_rmsd']]
        df_nplx['TM-score'] = df_nplx['TMScore']
        df_nplx['Backbone RMSD'] = df_nplx['Calpha_rmsd']
        df_nplx['Ligand RMSD'] = np.array(df_nplx['ligand_rmsd'], dtype=float)
        del df_nplx['TMScore']
        del df_nplx['Calpha_rmsd']
        del df_nplx['ligand_rmsd']
        df_nplx_summary = df_nplx.describe()
        df_nplx_summary['sampler'] = sampler_name[sampler]
        df_nplx_summary['Method'] = mode_name[mode]
        df_nplx_summary = df_nplx_summary[['Method', 'sampler', 'TM-score', 'Backbone RMSD', 'Ligand RMSD']]
        df_nplx_summary_top1 = df_nplx.sort_values('TM-score', ascending=False).drop_duplicates('sample_id').describe()
        df_nplx_summary = df_nplx_summary.join(df_nplx_summary_top1, lsuffix='\\n (all predictions)', rsuffix='\\n (per-target top-1)')
        df_nplx_summary = df_nplx_summary.drop(['count', 'min', 'max'])
        
        all_summary.append(df_nplx_summary)
        
all_summary = pd.concat(all_summary).fillna('-')
all_summary = all_summary.apply(pd.to_numeric, errors='ignore')
all_summary['stat'] = all_summary.index
all_summary = all_summary.set_index(['Method', 'sampler', 'stat'])
all_summary.to_csv('psp_summary_pocketminer.csv', float_format='%.3f')
pd.options.display.float_format = '{:,.3f}'.format
pretty_print(all_summary)
```
