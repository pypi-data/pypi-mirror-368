```python
import xarray as xr
import numpy as np
```

build a dummy array

```python
arr = np.random.rand(100,2,180)
```

turn it into an xarray dataset

```python
da = xr.DataArray(arr)
da
```

```python
ds = da.to_dataset(name='tl')
ds = ds.rename_dims({'dim_0':'sample','dim_1':'channel_id','dim_2':'time',})
ds
```

```python
ds.to_zarr('dummy_data.zarr', mode='w', consolidated=True)
```

```python
ds = xr.open_zarr('dummy_data.zarr')
```
